````python
# Quick setup verification
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import networkx as nx

print(f"Python version: {sys.version}")
print("‚úÖ All required packages imported successfully!")
print("\nüéØ Ready to explore data structures and algorithms!")
print("\nüìä Sample visualization test:")

# Simple complexity comparison visualization
n = np.linspace(1, 100, 100)
plt.figure(figsize=(10, 6))
plt.plot(n, n, label='O(n) - Linear', color='blue')
plt.plot(n, n * np.log(n), label='O(n log n) - Linearithmic', color='green')
plt.plot(n, n**2, label='O(n¬≤) - Quadratic', color='red')
plt.plot(n, 2**n, label='O(2^n) - Exponential', color='purple')
plt.yscale('log')
plt.xlabel('Input Size (n)')
plt.ylabel('Operations (log scale)')
plt.title('Algorithm Complexity Comparison')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## üìà Progress Tracking

Use this interactive widget to track your learning progress:

````python
import ipywidgets as widgets
from IPython.display import display, HTML

# Progress tracking widget
chapters = [
    "Chapter 1: Overview",
    "Chapter 2: Basic Python", 
    "Chapter 3: OOP",
    "Chapter 4: Testing",
    "Chapter 5: Complexity Analysis",
    "Chapter 6: Stacks & Queues",
    "Chapter 7: Deques & Linked Lists",
    "Chapter 8: Doubly Linked Lists",
    "Chapter 9: Recursion",
    "Chapter 10: Dynamic Programming",
    "Chapter 11: Binary Search",
    "Chapter 12: Sorting",
    "Chapter 13: Divide & Conquer Sorting",
    "Chapter 14: Selection",
    "Chapter 15: Hash Tables",
    "Chapter 16: Trees",
    "Chapter 17: BSTs",
    "Chapter 18: Balanced Trees",
    "Chapter 19: Priority Queues",
    "Chapter 20: Graphs",
    "Chapter 21: Advanced Graph Search",
    "Chapter 22: Union-Find"
]

progress_widgets = []
for chapter in chapters:
    checkbox = widgets.Checkbox(value=False, description=chapter)
    progress_widgets.append(checkbox)

progress_box = widgets.VBox(progress_widgets)
display(HTML("<h3>üìù Mark chapters as completed:</h3>"))
display(progress_box)

# Progress summary
def update_progress(*args):
    completed = sum(w.value for w in progress_widgets)
    total = len(progress_widgets)
    percentage = (completed / total) * 100
    print(f"\rüìä Progress: {completed}/{total} chapters completed ({percentage:.1f}%)", end='')

for widget in progress_widgets:
    widget.observe(update_progress, names='value')
    
update_progress()
````
